```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Sample - Superstore.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df2=df[['Order Date','Sales','Country/Region','Category','Profit']]
```

```python
df2.head()
```

```python
df2['Order Date']=pd.to_datetime(df2['Order Date'])
df2['Year']=df2['Order Date'].dt.year
```

```python
df2.dtypes
```

```python
print(df2['Year'].unique())
print(df2['Category'].unique())
print(df2['Country/Region'].unique())
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

# Year Wise Sale & Profit analysis

```python
sales_profit=df2.groupby("Year")[["Sales","Profit"]].sum()
print(sales_profit)
```

```python
sales_profit.plot(kind="bar")
plt.title("Year wise Sale Analysis")
```

# Time Series Analysis

```python
df2['Order Date']=pd.to_datetime(df2['Order Date'])
monthly=df2.groupby(pd.Grouper(key='Order Date',freq='ME'))['Sales'].sum()
monthly.plot(title='Monthly Sales Trend')
monthly.rolling(6).mean().plot(label='3-Month Avg')
```

# Country Wise Performance

```python
country_sales=df2.groupby('Country/Region')['Sales'].sum()
country_sales.sort_values().plot(kind='bar',title='Sales By Country')
```

# Which Category Profitable

```python
cat_profit=df2.groupby('Category')['Profit'].sum()
cat_profit.plot(kind='bar',title='Profit By Category')
```

```python
df2['Profit margin']=df2['Profit']/df2['Sales']*100
df2.head()
```

```python
df2.groupby('Category')['Profit margin'].mean().plot(kind='pie',title='Profit Margin By category')
```

# Yearly Sale Growth

```python
yearly=df2.groupby("Year")["Sales"].sum()
yearly.plot(marker='o',title='Yearly Sale Growth')
```

# Corelation Analysis

```python
import seaborn as sns
sns.heatmap(df2[['Sales','Profit']].corr(),annot=True)
```
